Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

 Seaborn is a powerful data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics.

Here are 5 commonly used Seaborn plots and their uses:

Line Plot (sns.lineplot)

Use: Shows trends over time or ordered categories.

Example: Tracking changes in temperature over days.

Bar Plot (sns.barplot)

Use: Displays the average value of a category.

Example: Average sales per product.

Box Plot (sns.boxplot)

Use: Shows distribution summary using quartiles and highlights outliers.

Example: Distribution of student scores.

Histogram (sns.histplot)

Use: Displays frequency distribution of a variable.

Example: Distribution of ages of people in a survey.

Pair Plot (sns.pairplot)

Use: Visualizes pairwise relationships in a dataset.

Example: Relationship between features in the iris dataset.



Q2: Load the "fmri" dataset and plot a line plot with x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset("fmri")

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Timepoint vs Signal")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.grid(True)
plt.show()
```

Q3: Load the "titanic" dataset and plot two box plots for age and fare grouped by pclass.

```python
# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot 1 - Age
plt.figure(figsize=(8, 4))
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Box Plot of Age by Pclass")
plt.show()

# Box plot 2 - Fare
plt.figure(figsize=(8, 4))
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot of Fare by Pclass")
plt.show()
```

Q4: Plot a histogram for the 'price' column in the "diamonds" dataset with hue='cut'.

```python
# Load diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x="price", hue="cut", element="step", stat="density", common_norm=False)
plt.title("Price Distribution by Cut")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```

Q5: Plot a pair plot of the "iris" dataset with hue = "species".

```python
# Load iris dataset
iris = sns.load_dataset("iris")

# Pair plot
sns.pairplot(data=iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

Q6: Plot a heatmap using the "flights" dataset.

```python
# Load flights dataset
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="coolwarm")
plt.title("Heatmap of Flights Dataset")
plt.show()
```
