Q2. Create a simple Flask application to display ‘Hello World!!’

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hellow world"

if __name__ == "__main__":
    app.run()
```

```python
# App Routing in Flask means mapping URLs to functions. When a user visits a specific URL, Flask uses the corresponding function to respond.

#  We use @app.route() to define routes.
#  It helps in organizing your application and creating multiple pages like /home, /about, /login, etc.
```

Q4. Create two routes: /welcome and /

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def company_info():
    return "Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999"


@app.route("/welcome")
def welcome():
    return "<h2>Welcome to ABC Corporation</h2>"

if __name__ == "__main__":
    app.run()
```

 Q5. What function is used in Flask for URL building?

```python
#Flask uses the url_for() function to dynamically build URLs using the name of the function, instead of hardcoding the route.

from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "Hello, World!"

@app.route("/profile")
def profile():
    return "This is your profile"

@app.route("/links")
def links():
    return f""" 
      
      <a href="{url_for('home')}">Home</a><br>
      <a href="{url_for('profile')}">Profile</a>
    
    """   
    
if __name__ == "__main__":
    app.run()         
```
