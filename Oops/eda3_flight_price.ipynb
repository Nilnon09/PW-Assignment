Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the

Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset
to identify the peak travel season. What features would you analyze to identify the peak season, and how
would you present your findings to your boss?

```python
flight_df['Departure Time'] = pd.to_datetime(flight_df['Departure Time'])
flight_df['Month'] = flight_df['Departure Time'].dt.month
print(flight_df['Month'].value_counts().sort_index())
```

Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?

```python
flight_df['DayOfWeek'] = flight_df['Departure Time'].dt.day_name()
flight_df['Hour'] = flight_df['Departure Time'].dt.hour

# Average price by day
print(flight_df.groupby('DayOfWeek')['Price'].mean().sort_values(ascending=False))

# Average price by hour
print(flight_df.groupby('Hour')['Price'].mean().sort_values(ascending=False))
```

Q8. You are a data scientist working for an airline company, and you have been asked to analyze the
Flight Price dataset to identify the factors that affect flight prices. What features would you analyze to
identify these factors, and how would you present your findings to the management team?

```python
# Encode categorical variables
encoded_df = pd.get_dummies(flight_df.drop(columns=['Departure Time']), drop_first=True)

# Correlation matrix
correlation_matrix = encoded_df.corr()
print(correlation_matrix['Price'].sort_values(ascending=False).head(10))
```

Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does
the dataset have?

```python
import pandas as pd

df = pd.read_csv("google_playstore_sample.csv")
df.shape
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.boxplot(x='Category', y='Rating', data=df)
plt.title('App Ratings by Category')
plt.xticks(rotation=45)
plt.show()
```

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
df.isnull().sum()
```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
# Convert 'Size' column from string to float
df['Size_MB'] = df['Size'].str.replace('M', '').astype(float)

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Size_MB', y='Rating', data=df)
plt.title('App Size vs Rating')
plt.xlabel('Size (MB)')
plt.ylabel('Rating')
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
avg_price = df.groupby('Type')['Price'].mean().reset_index()

sns.barplot(x='Type', y='Price', data=avg_price)
plt.title('Average App Price by Type')
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
# Clean 'Installs' column and convert to int
df['Installs_clean'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
top_apps = df.sort_values(by='Installs_clean', ascending=False)[['App', 'Installs']]
top_apps.head(10)
```

Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to
launch a new app. What features would you analyze to make recommendations to the company, and
what data visualizations would you use to present your findings?

To identify the best time to launch a new app, we’d need to analyze temporal patterns in app performance. Since the dataset we created doesn't include temporal features, let's extend it by adding a "Last Updated" field and simulate analysis based on it.

Features to Analyze:

Last Updated – To understand when apps were last maintained or launched.

Installs – To correlate app launch/update timing with popularity.

Rating – To observe whether newer apps tend to have better engagement.

```python
df['Last Updated'] = pd.to_datetime([
    '2023-01-15', '2023-06-20', '2022-12-05', '2023-03-30', '2023-07-10',
    '2022-11-01', '2023-04-12', '2023-08-25', '2023-02-18', '2023-05-01'
])
```

```python
df['Updated Month'] = df['Last Updated'].dt.month_name()

sns.boxplot(x='Updated Month', y='Installs_clean', data=df)
plt.title('Installs by Update Month')
plt.xticks(rotation=45)
plt.ylabel('Installs')
plt.show()
```

```python
df['Days Since Update'] = (pd.Timestamp('2025-04-18') - df['Last Updated']).dt.days

sns.scatterplot(x='Days Since Update', y='Installs_clean', data=df)
plt.title('App Recency vs Installs')
plt.xlabel('Days Since Last Update')
plt.ylabel('Installs')
plt.show()
```
