Q2: What is a scatter plot? Plot a scatter plot with the given code.

 A scatter plot is a type of plot that shows the relationship between two variables using dots. It's useful for identifying correlations, patterns, and outliers.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using subplot().

The subplot() function is used to display multiple plots in one figure. It divides the plotting area into a grid and allows plotting in specific sub-regions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Line data
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'r')
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g')
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'm')
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Plot a bar and horizontal bar plot using given data.

A bar plot displays data using rectangular bars. It’s useful to compare quantities across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='skyblue')
plt.title("Company Profit - Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='orange')
plt.title("Company Profit - Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Plot a box plot using given data.

A box plot (or box-and-whisker plot) shows the distribution of data based on five-number summary: min, Q1, median, Q3, and max. It helps in identifying outliers and data spread.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
